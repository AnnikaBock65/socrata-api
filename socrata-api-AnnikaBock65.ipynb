```python
df_list = []

for num in range(0,13):
    endpoint = 'https://data.nashville.gov/resource/7qhx-rexh.json'
    params = {'$limit':50000, '$offset':num*50000}
    response = requests.get(endpoint, params = params)
    res = response.json()
    df_list.extend(res)
```
